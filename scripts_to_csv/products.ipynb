```python
import pandas as pd
import numpy as np
from faker import Faker
from random import randint
from datetime import timedelta
import datetime
import requests
from bs4 import BeautifulSoup
import random
```

```python
id = [i for i in range(1, 1489)]
```

```python
category_id = [i for i in range(1, 4)]
```

```python
clothes = []
for i in range(1, 50):
  url = 'https://www.lamoda.ru/c/477/clothes-muzhskaya-odezhda/?sitelink=topmenuM&l=3&page=' + str(i)
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  items = soup.find_all('div', class_='x-product-card__card')
  for item in items:
    item = item.find('div', class_='x-product-card-description__product-name').text
    if item not in clothes:
      clothes.append(item)
    else:
      continue
```

```python
shoes = []
for i in range(1, 50):
  url = 'https://www.lamoda.ru/c/17/shoes-men/?sitelink=topmenuM&l=4&page=' + str(i)
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  items = soup.find_all('div', class_='x-product-card__card')
  for item in items:
    item = item.find('div', class_='x-product-card-description__product-name').text
    if item not in shoes:
      shoes.append(item)
    else:
      continue
```

```python
accessories = []
for i in range(1, 50):
  url = 'https://www.lamoda.ru/c/559/accs-muzhskieaksessuary/?sitelink=topmenuM&l=5&page=' + str(i)
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  items = soup.find_all('div', class_='x-product-card__card')
  for item in items:
    item = item.find('div', class_='x-product-card-description__product-name').text
    if item not in accessories:
      accessories.append(item)
    else:
      continue
```

```python
len(clothes) + len(shoes) + len(accessories)
```

```python
cost = [
    random.randint(299, 30000) for _ in range(1, 1489)
]
```

```python
print(len(clothes))
print(len(shoes))
print(len(accessories))
```

```python
category1 = [1] * 496
category2 = [2] * 496
category3 = [3] * 496

category_id = category1 + category2 + category3
name = clothes + shoes + accessories
```

```python
d = {
    'category_id': category_id,
     'name': name
}
```

```python
df = pd.DataFrame(d)
```

```python
df['cost'] = cost
```
