```python
id = [i for i in range(1, 401)]
```

```python
client_id = [
    random.randint(1, 150) for _ in range(1, 401)    
]
```

```python
fake = Faker(locale="ru_RU")
date_of_registration = [
    fake.date_between(start_date='-3y', end_date='now') for _ in range(1, 401)
]
```

```python
choice = ['OK', 'NOT OK']
order_status = [
    random.choice(choice) for _ in range(1, 401)
]
```

```python
df = pd.DataFrame(
    {
        'id': id,
        'client_id': client_id,
        'date_of_registration': date_of_registration,
        'order_status': order_status
    },
    index = [i for i in range(1, 401)]
)
```

```python
df['completion_date'] = np.where(df.order_status == 'OK', df.date_of_registration + timedelta(random.randint(1, 2)), '')
```

```python
df
```

```python
products = pd.read_csv('/content/products.csv')
```

```python
ordering_products = pd.read_csv('/content/ordering_products.csv')
```

```python
products = products.drop(columns='Unnamed: 0')
ordering_products = ordering_products.drop(columns='Unnamed: 0')
```

```python
res = ordering_products.merge(products, left_on='product_id', right_on='id')
```

```python
res
```

```python
result = res.merge(df, left_on='order_id', right_on='id')
```

```python
result.cost = result.cost.apply(lambda x: int(str(x)[0:len(x) - 4]))
```

```python
result
```

```python
result
```

```python
result['order_value'] = np.where(result.order_status == 'OK', result.quantity * result.cost, 0)
```

```python
orders = result.drop(columns=['Unnamed: 0_x', 'id_x', 'order_id', 'product_id', 'quantity', 'Unnamed: 0_y', 'id_y', 'category_id', 'name', 'cost'])
```
