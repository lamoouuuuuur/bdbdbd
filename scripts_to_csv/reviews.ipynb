```python
import pandas as pd
import numpy as np
import random
```

```python
client = pd.read_csv(r'/content/сlient.csv')
```

```python
products = pd.read_csv(r'/content/products.csv')
```

```python
client = client.drop(columns='Unnamed: 0')
```

```python
products = products.drop(columns='Unnamed: 0')
```

```python
id = [i for i in range(1, 401)]
```

```python
rating = [1, 2, 3, 4, 5]

rate = [random.choice(rating) for _ in range(1, 401)]
```

```python
product_id = [
    random.randint(1, 1489) for _ in range(1, 401)
]
```

```python
client_id = [
    random.randint(1, 149) for _ in range(1, 401)
]
```

```python
reviews = pd.DataFrame(
    {
        'id': id, 
        'client_id': client_id,
        'product_id': product_id,
        'rate': rate
    },
    index = [i for i in range(1, 401)]
)
```

```python
reviews
```

```python
res = reviews.merge(client, left_on='client_id', right_on='id')
```

```python
res
```

```python
result = res.merge(products, left_on='product_id', right_on='id')
```
