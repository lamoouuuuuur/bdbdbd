```python
import pandas as pd
import numpy as np
import random
```

```python
brand = pd.read_csv(r'/content/brand.csv')
color = pd.read_csv(r'/content/color.csv')
size = pd.read_csv(r'/content/size.csv')
products = pd.read_csv(r'/content/products.csv')
```

```python
brand = brand.drop(columns='Unnamed: 0')
color = color.drop(columns='Unnamed: 0')
size = size.drop(columns='Unnamed: 0')
products = products.drop(columns='Unnamed: 0')
```

```python
size_shoes_1 = size.iloc[7:].to_numpy()
size_clothes_1 = size.iloc[0:7].to_numpy()
```

```python
dict_size_shoes = {
    size_shoes_1[i][0]: size_shoes_1[i] for i in range(len(size_shoes_1))
}

size_shoes = list(dict_size_shoes.keys())

dict_size_clothes = {
    size_clothes_1[i][0]: size_clothes_1[i] for i in range(len(size_clothes_1))
}

size_clothes = list(dict_size_clothes.keys())
```

```python
id = [
    i for i in range(1, 1489)
]

brand_id = [
    random.randint(1, 61) for _ in range(1, 1489)
]

color_id = [
    random.randint(1, 21) for _ in range(1, 1489)
]

product_id = [
    random.randint(1, 1001) for _ in range(1, 1489)
]

  #   size_id.append(random.choice(list(dict_size_shoes.keys())))
  # else:
  #   size_id.append(random.choice(list(dict_size_clothes.keys())))
```

```python
df = pd.DataFrame(
    {
        'id': id,
        'product_id': product_id,
        'color_id': color_id,
        'brand_id': brand_id
    },
    index = [i for i in range(1, 1489)]
)
```

```python
df1 = df.merge(products, left_on='product_id', right_on='id')
```

```python
df1
```

```python
for_category_2 = [random.choice(size_shoes) for _ in range(1, 1489)]
for_category_else = [random.choice(size_clothes) for _ in range(1, 1489)]

df1['size_id'] = np.where(df1.category_id == 2, for_category_2, for_category_else)
```
