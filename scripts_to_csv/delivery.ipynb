```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta
import datetime
```

```python
products = pd.read_csv(r'/content/products.csv')
```

```python
products = products.drop(columns='Unnamed: 0')
```

```python
pickup_point_df = pd.read_csv(r'/content/pickup_point.csv')
```

```python
pickup_point_df = pickup_point_df.drop(columns='Unnamed: 0')
```

```python
pickup_point_df.head(2)
```

```python
pickup_point_df['time_start'] = pd.to_datetime(pickup_point_df['time_start'])
pickup_point_df['time_end'] = pd.to_datetime(pickup_point_df['time_end'])
```

```python
pickup_point_df
```

```python
id = [
    i for i in range(1, 101)
]

order_id = [
    random.randint(1, 401) for _ in range(1, 101)
]

pickup_point = [
    random.randint(1, 101) for _ in range(1, 101)
]

lst_cost = products.cost.apply(lambda x: int(str(x)[0:len(x) - 4])).to_numpy()

delivery_cost = [ 
    lst_cost[i] + random.randint(500, 1000) for i in range(1, 101)
]


lst = (pickup_point_df.time_end - pickup_point_df.time_start).dt.days.to_numpy()

date_delivery = [
    lst[i] for i in range(len(lst))
]
```

```python
delivery = pd.DataFrame(
    {
        'id': id,
        'order_id': order_id,
        'pickup_point': pickup_point,
        'delivery_cost': delivery_cost,
        'date_delivery': date_delivery
    },
    index = [i for i in range(1, 101)]
)
```

```python
delivery
```

```python
delivery.to_csv('delivery.csv')
```
