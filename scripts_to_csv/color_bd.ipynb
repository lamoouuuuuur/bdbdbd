```python
from faker import Faker
import pandas as pd
```

```python
fake = Faker('ru-RU')
```

```python
id = [i for i in range(1, 50)]
```

```python
name = [fake.color_name() for _ in range(1, 151)]
```

```python
name = {
    name[i]: name.count(name[i]) for i in range(len(name))
}
```

```python
name
```

```python
color = pd.DataFrame(
    {
        'id': id,
        'name': name.keys()
    },
    index=[i for i in range(1, 50)]
)
```

```python
color
```

```python
color.to_csv('color.csv')
```
